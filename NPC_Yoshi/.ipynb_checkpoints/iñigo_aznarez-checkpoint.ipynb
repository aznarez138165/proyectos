* Relación con Otros Personajes -> Estado de ánimo: La relación de Yoshi con otros personajes podría influir en su

## 6. La programación en Python de la Red Bayesiana que has creado.
 

```python
import random
from collections import deque

class Nodo:
    def __init__(self, padres, hijos, p):
        self.padres = padres
        self.hijos = hijos
        self.p = p
        self.num_states = 2
        self.states = [0, 1]
        self.state = None
    
    def sample(self, estado_padres):
        # Genera numero aleatorio
        n = random.random()

        # nodo sin padres
        if len(estado_padres) == 0:
            if n <= self.p[0][0]:
                self.state = 0
            else:
                self.state = 1    

        # nodo con un solo padre
        elif len(estado_padres) == 1:
            if n <= self.p[estado_padres[0]][0]:
                self.state = 0
            else:
                self.state = 1        

        # nodo con dos padres
        elif len(estado_padres) == 2:
            if estado_padres[0] is not None and estado_padres[1] is not None :
                if n <= self.p[estado_padres[0]][estado_padres[1]][0]:
                    self.state = 0
                else:
                    self.state = 1   

        # nodo con tres padres
        elif len(estado_padres) == 3:
            if estado_padres[0] is not None and estado_padres[1] is not None and estado_padres[2] is not None:
                if n <= self.p[estado_padres[0]][estado_padres[1]][estado_padres[2]][0]:
                    self.state = 0
                else:
                    self.state = 1
        

# Definir los nodos de la Red Bayesiana
# 0. Dificultad
p_dificultad = [[0.6, 0.4]]
dificultad = Nodo([], [5], p_dificultad)

# 1. Tiempo
p_tiempo = [[0.7, 0.3]]
tiempo = Nodo([], [6], p_tiempo)

# 2. Objetivo
p_objetivo = [[0.5, 0.5]]
objetivo = Nodo([], [6], p_objetivo)

# 3. Relación con otros personajes
p_relacion = [[0.7, 0.3]]
relacion = Nodo([], [9], p_relacion)

# 4. Inventario
p_inventario = [[0.4, 0.6]]
inventario = Nodo([], [8], p_inventario)

# 5. Peligro
p_peligro = [[0.3, 0.7], [0.1, 0.9]]
peligro = Nodo([0], [6], p_peligro)

# 6. Huevos
p_huevos = [[[[0.8, 0.2], [0.5, 0.5]], [[0.6, 0.4], [0.3, 0.7]]], [[[0.2, 0.8], [0.1, 0.9]], [[0.4, 0.6], [0.05, 0.95]]]]
huevos = Nodo([1, 2, 5], [7, 8], p_huevos)

# 7. Hambre
p_hambre = [[0.1, 0.9], [0.8, 0.2]]
hambre = Nodo([6], [9], p_hambre)

# 8. Salud
p_salud = [[[0.7, 0.3], [0.4, 0.6]], [[0.3, 0.7], [0.1, 0.9]]]
salud = Nodo([6, 4], [], p_salud)

# 9. Estado de Ánimo
p_estado_animo = [[[0.6, 0.4], [0.8, 0.2]], [[0.1, 0.9], [0.6, 0.4]]]
estado_animo = Nodo([3, 7], [], p_estado_animo)

# Construir el grafo
grafo = {
    0: dificultad,
    1: tiempo,
    2: objetivo,
    3: relacion,
    4: inventario,
    5: peligro,
    6: huevos,
    7: hambre,
    8: salud,
    9: estado_animo
}
```

```python
#funciones 

# imprime: imprime los estados de los nodos del grafo
def imprime(grafo):
    estados={}
    for nodo in grafo.keys():
        estados[nodo] = grafo[nodo].state
    print(estados)

# establece todos los estados de los nodos del grafo a None
def reset_grafo (grafo):
    for nodo in grafo.keys():
        grafo[nodo].state = None
    return grafo

# comprueba si un nodo está en una lista de nodos
def estaen(lista, nodo):
    for nodos in lista:
        if nodos == nodo:
            return True
    return False

# recorre en anchura el grafo dando valores a los estados
def recorrido_anchura(node, graph):
    abiertos = deque()
    cerrados = []

    for key in graph:
        if len(graph[key].padres) == 0:
            abiertos.append(graph[key])
        
    while(len(abiertos) > 0):
        estados = []
        actual = abiertos[0]
        abiertos.popleft()
        cerrados.append(actual)
        for i in actual.padres:
            estados.append(graph[i].state)

        actual.sample(estados)

        for hijo in actual.hijos:
            if not estaen(abiertos, graph[hijo]) and not estaen(cerrados, graph[hijo]):
                abiertos.append(graph[hijo])
            
    return graph
```

## 7. Utilizando el método de muestreo con rechazo, programar dos casos prácticos de funcionamiento de la red, es decir programar dos casos en los que las evidencias y las queries sean diferentes.

```python
# crea una muestra aleatoria del grafo
def sample_grafo(node, graph):
    # resetear el grafo
    reset_grafo(graph)
    # realizar el recorrido en anchura donde se crea la muestra aleatoria
    recorrido_anchura(node, graph)
    # devolver los estados de los nodos del grafo
    estados = {}
    for nodo in graph:
        estados[nodo] = graph[nodo].state
    return estados

# dadas una query, una evidence, un grafo y un número de muestras, la inferencia realiza el muestreo con rechazo   
def infer(query, evidence, graph, num=20000):
    n_evidence = 0
    n_query_and_evidence = 0

    # sacar el nodo raiz
    keys = []
    for key in grafo.keys():
        keys.append(key)

    i = keys[0]
    nodo_raiz = graph[i]
    
    # muestreo
    for _ in range(num):
        # generar muestras aleatorias de los estados de las variables en el grafo
        estados = sample_grafo(nodo_raiz, graph) 
        #imprime(graph)
        
        # comprobar si la muestra satisface la evidencia
        for key, value in evidence.items():
            # Obtener el valor del estado del nodo correspondiente a la clave en la muestra
            estado = estados[key]
            # Comprobar si el valor del estado coincide con el valor de la evidencia
            if estado == value:
                # coincide, continuar con la siguiente condición
                continue
            else:
                # no coincide, la condición de no se cumple
                evidencia_valida = False
                break
        else:
            # todas las condiciones de evidencia se cumplen
            evidencia_valida = True

        if evidencia_valida:
            n_evidence += 1

            # comprobar si la muestra satisface tanto la evidencia como la consulta
            for key, value in query.items():
                # Obtener el valor del estado del nodo correspondiente a la clave en la muestra
                estado = estados[key]
                # Comprobar si el valor del estado coincide con el valor de la evidencia
                if estado == value:
                    # coincide, continuar con la siguiente condición
                    continue
                else:
                    # no coincide, la condición de no se cumple
                    evidencia_consulta_valida = False
                    break
            else:
                # todas las condiciones de evidencia se cumplen
                evidencia_consulta_valida = True

            if evidencia_consulta_valida:
                n_query_and_evidence += 1

    # Calcular la probabilidad condicional aproximada
    print("Evidence:", n_evidence)
    print("Query y evidence:", n_query_and_evidence)
    if n_evidence == 0:
        return 0.0
    else:
        return n_query_and_evidence / n_evidence
```

## Caso práctico 1
Supongamos que queremos calcular la probabilidad de que Yoshi esté herido dado que el objetivo es derrotar al jefe final, el nivel sea fácil y no haya tiempo límite.
Definimos las evidencias y la consulta:
* Evidencia = el objetivo es derrotar al jefe final, el nivel sea fácil y no hay tiempo límite
* Consulta = Yoshi está herido

```python
evidencia = {2:1, 0:0, 1:1}  # el objetivo es derrotar al jefe final, el nivel sea difícil y no hay tiempo límite
consulta = {8: 0}   # Yoshi esta herido

# Realizamos la inferencia
probabilidad = infer(consulta, evidencia, grafo)
print("La probabilidad de que Yoshi esté herido dado que el objetivo es derrotar al jefe final, el nivel sea fácil y no haya tiempo límite es:", probabilidad)
```
